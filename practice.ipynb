```python
import numpy as np;
import pandas as pd;
import matplotlib.pyplot as plt
```

```python
cols=['sepal length','sepal width','petal length','petal width','class'];
df=pd.read_csv('iris.data',names=cols);
class_mapping = {'Iris-setosa': 0, 'Iris-versicolor': 1, 'Iris-virginica': 2}
df['class'] = df['class'].map(class_mapping)
print(df.head());
```

```python
print(df['class'].unique());
```

```python
for label in cols[:-1]:
  plt.hist(df[df['class']==0][label], color='blue',label='alpha',alpha=0.7,density=True);
  plt.hist(df[df['class']==1][label], color='red',label='beta',alpha=0.7,density=True);
  plt.hist(df[df['class']==2][label], color='green',label='gamma',alpha=0.7,density=True);
  plt.title(label)
  plt.ylabel('Probability');
  plt.xlabel('label');
  plt.legend();
  plt.show();

```
